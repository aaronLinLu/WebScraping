## 1. Statistics in Presidential Debates

Your first task is to scrape Presidential Debates from the Commission of Presidential Debates website: http://www.debates.org/index.php?page=debate-transcripts.

To do this, you are not allowed to manually look up the URLs that you need, instead you have to scrape them. The root url to be scraped is the one listed above, namely: http://www.debates.org/index.php?page=debate-transcripts


1. By using `requests` and `BeautifulSoup` find all the links / URLs on the website that links to transcriptions of **First Presidential Debates** from the years [2012, 2008, 2004, 2000, 1996, 1988, 1984, 1976, 1960]. In total you should find 9 links / URLs tat fulfill this criteria.
2. When you have a list of the URLs your task is to create a Data Frame with some statistics (see example of output below):
    1. Scrape the title of each link and use that as the column name in your Data Frame. 
    2. Count how long the transcript of the debate is (as in the number of characters in transcription string). Feel free to include `\` characters in your count, but remove any breakline characters, i.e. `\n`. You will get credit if your count is +/- 10% from our result.
    3. Count how many times the word **war** was used in the different debates. Note that you have to convert the text in a smart way (to not count the word **warranty** for example, but counting **war.**, **war!**, **war,** or **War** etc.

````python
from IPython.core.display import display, HTML
from __future__ import division, print_function
import requests
import bs4 as bs
from collections import Counter
import re
import pandas as pd

source = requests.get('http://www.debates.org/index.php?page=debate-transcripts')
soup = bs.BeautifulSoup(source.content, features='html.parser') 

tab = [] # all the 9 titles
li = [] # all the links for the 9 debates
links = soup.find_all('a') # all the links in the website

for l in links:
    if ('First' in l.text) & ('Presidential' in l.text) & ('Debate' in l.text):
        #print('Info about {}: '.format(l.text), l.get('href'))
        tab.append(l.text)
        li.append(l.get('href'))
````

````python
#### 2A. Scrape the title of each link and use that as the column name in your Data Frame.
df_stats = pd.DataFrame(columns=tab,index=['Debate char length','war_count','most_common_w','most_common_w_count'])
````

````python
for i in range(len(tab)):
    # locate current debate
    current_debate = tab[i]
    
    length, war = [],[]
    # look at each transcript.... 
    co = requests.get(li[i])
    cont = bs.BeautifulSoup(co.content,features='html.parser')
    # find all p tags in the html element
    x = cont.find_all('p')
    # b: main transcript 
    b = x[0]
    #### 2B. Count how long the transcript of the debate is
    ll = 0
    co_split = Counter(re.findall('[0-9A-Za-z?.,;!-"'' ]', b.text.lower()))
    for j in range(len(co_split.most_common())):
        ll += co_split.most_common()[j][1]
    #### 2C. Count how many times the word war was used in the different debates
    war_len = 0
    for ele in x:
        war_vars = ['War,','war,','Wars','wars','War.','war.',
                    'War!','war!','War:','war:']
        for w in war_vars:
            war_len += ele.text.count(w)
        war.append(war_len)
    #### 2D. scrape the most common used word in the debate, and write how many times it was used
    counts = Counter(re.findall(r"[\w]+",b.text.lower()))
    ## append the result
    df_stats[tab[i]] = [ll,war_len,counts.most_common()[0][0],counts.most_common()[0][1]]
````

Scrape the first 27 data sets from this URL http://people.sc.fsu.edu/~jburkardt/datasets/regression/ (i.e.`x01.txt` - `x27.txt`). Then, save the 5th line in each data set, this should be the name of the data set author (get rid of the `#` symbol, the white spaces and the comma at the end). 

Count how many times (with a Python function) each author is the reference for one of the 27 data sets. Showcase your results, sorted, with the most common author name first and how many times he appeared in data sets. Use a Pandas DataFrame to show your results, see example.

````python
# locate our targeted page
source_url = "http://people.sc.fsu.edu/~jburkardt/datasets/regression/"
source = requests.get(source_url)
soup = bs.BeautifulSoup(source.content,features='html.parser')
````

````python
authors = []
for n, link in enumerate(soup.find_all("a")): # grab url links
    try:
        if '28' in link.text:
            break
        elif 'txt' in link.text:
            file_url = source_url+link.get('href')
            file_name = link.get('href')
            file_source = requests.get(file_url,stream=True)
            file_soup = bs.BeautifulSoup(file_source.content,features='lxml')
            # locate passage body
            x = file_soup.find_all('p')
            for l in x:
                text = l.text.split("\n")
            author = text[4].strip('# ,')
            authors.append(author)
            del file_source
    except (TimeoutError,ConnectionError):
        print('connected party failed to respond.')
````

````python
# customized function to return arthor name, and his/her times of references, formatted in a dataframe
def CountRefTimes(authors_list):
    # grab necessary libraries
    from collections import OrderedDict
    from operator import itemgetter
    import pandas as pd
    # processing
    d = dict((x,authors_list.count(x)) for x in set(authors_list))
    # sort values (counts) in descending order
    d2 = [(k, d[k]) for k in sorted(d, key=d.get, reverse=True)]
    authors,counts = [],[]
    for k, v in d2:
        authors.append(k)
        counts.append(v)
    # create an empty dataframe to return
    df = pd.DataFrame(columns=['Authors','Counts'])
    df['Authors'] = authors
    df['Counts'] = counts
    # return it
    return df
````

````python
df_authorCts = CountRefTimes(authors_list=authors)
df_authorCts
````
